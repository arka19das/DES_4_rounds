```
#plain_text in hexadecimal
def encrypt(plain_text,sub_keys):
  plain_textb=""
  for x in range(16):
    plain_textb+=mp[plain_text[x]]
  plain_textbp=permutation(plain_textb)  #Intital permutation
  #plain_textp = plain_textb                   #Intital permutation removed
  L,R=plain_textbp[0:32],plain_textbp[32:64] #32 bits
  for i in range(4): #changeno
    out=func(R,sub_keys[i]) #function called on right half data
    temp=int(out,base=2)^int(L,base=2)
    L=R
    R='{0:032b}'.format(temp)
  final = inv_permutation(L+R)  
  #Inverse permutation after 4 rounds
  #final = left+right             #Inverse permutation removed
  cipher = hex(int(final,2))[2:]
  return cipher
  


def enc(plain_text,key):
  key_64=key  #has 16 character hexadecimal string
  sub_keys=gen_4_round_keys(key_64)
  return encrypt(plain_text,sub_keys)  
```
